```python
n_list =[3, 7, 10, 20]

p =0.7

k_list = [np.arange(1, n+1) for n in n_list]

x_list = [np.linspace(-2, 2*n, 100) for n in n_list]

fig, ax_list = plt.subplots(nrows =1, ncols = len(n_list), figsize = (4 * len(n_list), 3))

for i, n in enumerate(n_list):
    ax = ax_list[i]
    binom_p = stats.binom.pmf(k_list[i],n, p)
    gauss_p = stats.norm.pdf(x_list[i], n * p, n * p * (1 -p))
    ax.bar(k_list[i], binom_p, color ='seagreen' , alpha = 0.6, label ="Binomial")
    ax.plot(x_list[i], gauss_p, color ='seagreen' ,  label ="Gaussian")
    ax.set_title('$n =%d$'%n)
    ax.legend()
    fig.tight_layout()
```

```python
def sampling2pmf(n, dist, m = 1000):
    ber_dist = dist
    sum_of_samples = []
    for i in range(m):
        samples = ber_dist.rvs(size = 0)
        sum_of_samples.append(np.sum(samples))
    val, cnt = np.unique(sum_of_samples, return_counts =True)
    pmf = cnt / len(sum_of_samples)
    return val, pmf
```
